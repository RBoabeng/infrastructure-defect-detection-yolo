```python
import os
import cv2
import yaml
import matplotlib.pyplot as plt

project_root = "../" 
yaml_path = None
dataset_path = None

print("Searching for the TRUE YOLO dataset...")
for root, dirs, files in os.walk(project_root):
    # Skip virtual environments, git, and config folders so we don't get tricked again!
    if 'venv' in root or '.git' in root or 'config' in root: 
        continue
        
    # A true YOLO dataset will ALWAYS have a 'data.yaml' AND a 'train' folder together
    if 'data.yaml' in files and 'train' in dirs:
        yaml_path = os.path.join(root, 'data.yaml')
        dataset_path = root
        break

if yaml_path:
    print(f"✅ Successfully found the TRUE dataset hiding at: {dataset_path}")
    
    # Read the data.yaml file to get our defect class names
    with open(yaml_path, 'r') as file:
        data_info = yaml.safe_load(file)
        
    if isinstance(data_info['names'], dict):
        classes = list(data_info['names'].values())
    else:
        classes = data_info['names']
        
    print(f"Defect Classes in this dataset: {classes}")
else:
    print("❌ Could not find the dataset!")
    print("WARNING: Are you absolutely sure you ran `python download_data.py` on this specific school computer?")

```
